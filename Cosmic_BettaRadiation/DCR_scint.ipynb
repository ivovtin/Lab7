```python
import array
from array import array
import matplotlib.pyplot as plt
```

```python
##Draw DCR

##set up the correct directory name with data
path = "DCR"
##set up the correct file name with data
fileName1 = "wo_scint_53,92V_40db_gate100ms_staircase.txt"
fileName2 = "with_scint_53,92V_40db_gate100ms_staircase.txt"
fileName3 = "Eu_with_scint_53,92V_40db_gate100ms_staircase.txt"
fileName4 = "Sr_with_scint_53,92V_40db_gate100ms_2_staircase.txt"

listFiles = [fileName1,fileName2,fileName3,fileName4] 

outFile = 'DCR_sipm.png'
    
for fileName in listFiles:
    file = open(path + "/" + fileName)
    trhU, freq = array( 'd' ), array( 'd' )
    for line in file.readlines():
        line = line.replace(',','.')
        b = line.split()
        trhU.append(float(b[0]))
        freq.append(float(b[1]))
    
    plt.plot(trhU,freq, 'o', ls='-', ms=4)

plt.grid(True)
plt.yscale('log')
plt.xlabel('Threshold, mV')
plt.ylabel('Frequency, Hz')
plt.title('DCR')

plt.savefig(outFile )
plt.show()
```

```python
import array
from array import array
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy import stats
from scipy.signal import find_peaks
import numpy as np
from math import sqrt
```

```python
##set up the correct directory name with data
path = "ADCspectr"

##Cosmic
fileName1 = "Cosmic_53,92V_40dB_trh35mV_histo.txt"
##Eu
fileName2 = "Eu_53,92V_40dB_trh35mV_histo.txt"
##Sr
fileName3 = "Sr_53,92V_40dB_trh35mV_2_histo.txt"

listFiles = [fileName1,fileName2,fileName3]

outSpectr = 'spectr.png'  
```

```python
for fileName in listFiles:    
    x,y = np.array([]), np.array([])
    
    file1 = open(path + "/" + fileName)
    for line1 in file1.readlines():
        line1 = line1.replace(',','.')
        b = line1.split()
        b[0] = float(b[0])
        b[1] = float(b[1])
        x = np.append(x, b[0])
        y = np.append(y, b[1])
    
    #plt.plot(x, y,'o',markersize=4)  
    plt.bar(x, y, width = 10.0)  
    plt.xlabel('ADC channels')
    plt.ylabel('Counts')
    plt.savefig(str(listFiles.index(fileName)) + outSpectr)    
    plt.show() 
```

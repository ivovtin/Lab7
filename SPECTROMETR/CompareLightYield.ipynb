```python
import array
from array import array
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy import stats
from scipy.signal import find_peaks
import numpy as np
import math
from math import sqrt
```

```python
##set up the correct directory name with data
path = "CalScale30032023"

fileName1 = "53,98V_40dB_trh40_LYSO_histo.txt"
fileName2 = "53,98V_40dB_trh40_CsI_histo.txt"
fileName3 = "53,98V_40dB_trh40_BGO_histo.txt"

listFiles = [fileName1,fileName2,fileName3]

outAbsLaw = 'CompareLightYield.png'

XZoom = True
Xleft = 228000.0  
Xright = 242800.0  

YZoom = False
Ytop = 6000.0  
Ybottom = 0.0  

Peak1 = np.array([])
```

```python
for fileName in listFiles:    
    x,y = np.array([]), np.array([])
    
    try:
        file = open(path + "/" + fileName)
        for line in file.readlines():
            line = line.replace(',','.')
            b = line.split()
            b[0] = float(b[0])
            b[1] = float(b[1])
            x = np.append(x, b[0])
            y = np.append(y, b[1])
    except FileNotFoundError:
        print('Can`t open the file',fileName,'. No such file or directory')
    except ValueError:
        print('It`s not a number. Check the data in the file')
    except IndexError:
        print('The index is not in the range of elements. Check the data in the file')

    plt.plot(x, y,'o',markersize=2)
        
plt.xlabel('ADC channels')
plt.ylabel('Counts')
plt.yscale('log')
if XZoom:
    plt.xlim([Xleft, Xright])
if YZoom:
    plt.ylim([Ybottom, Ytop])
plt.show() 
```
